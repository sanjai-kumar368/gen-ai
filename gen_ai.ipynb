```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load base T5 model and tokenizer
model = T5ForConditionalGeneration.from_pretrained('t5-base')
tokenizer = T5Tokenizer.from_pretrained('t5-base')

def t5_task(input_text, max_length=100):
    # Tokenize and encode input
    input_ids = tokenizer(input_text, return_tensors="pt").input_ids
    # Generate output
    output_ids = model.generate(input_ids, max_length=max_length, num_beams=4, early_stopping=True)
    # Decode output
    return tokenizer.decode(output_ids[0], skip_special_tokens=True)

# Test all tasks
examples = {
    "Summarization": "summarize: India is a country with rich heritage and diversity. It has 28 states and 8 union territories.",
    "Translation": "translate English to French: I love machine learning.",
    "Grammar Correction": "grammar: He go to school yesterday.",
    "Question Generation": "generate question: Sachin Tendulkar is a famous Indian cricketer.",
    "Paraphrasing": "paraphrase: The weather is very pleasant today.",
    "Text Classification": "classify sentiment: I am extremely happy today!",
    "Headline Generation": "generate headline: Apple unveils new iPhone with powerful AI features and sleek design.",
    "Fill-in-the-blank": "fill: The capital of France is <extra_id_0>.",
    "Chatbot Response": "chat: What's your favorite movie?",
    "Explain or Elaborate": "explain: Newton's First Law"
}

# Run each task
for task, prompt in examples.items():
    result = t5_task(prompt)
    print(f"{task}: {result}")
```
